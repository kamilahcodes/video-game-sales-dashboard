```python
# imports 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('~/Desktop/Dev/data/vgsales.csv')
```

```python
# explore data set
df.info()
```

```python
df.dtypes
```

```python
# find null values: 
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Convert float year to integer, then to string
df['Year'] = df['Year'].astype(int).astype(str)
```

```python
# Parse string year as datetime and extract only the year
df['Year'] = pd.to_datetime(df['Year']).dt.year
```
